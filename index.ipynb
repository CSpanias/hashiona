## General Information
1. The thyroid is an endocrine gland found in the neck whose function is to **produce hormones (FT3 and FT4)** which it releases into the bloodstream$^4$.

<p style="text-align: center;">&#8595;</p>

2. **Thyroid hormones regulate**, among other things, **heart rate**, **body temperature**, and above all **metabolism**, that is the way the body uses and consumes nutrients$^4$. 

<p style="text-align: center;">&#8595;</p>

3. The thyroid gland may function more than normal (hyperthyroidism with increased hormones) or **less than normal** (**hypothyroidism with low hormones**) and in both cases, **major disorders** can occur$^4$. 

<p style="text-align: center;">&#8595;</p>

4. Furthermore, the thyroid gland can undergo **inﬂammation** (**thyroiditis**) or enlarge due to one or more swellings that are formed inside it (nodules, multinodular goiter). Some of these nodules can be the site of **malignant tumors**$^4$.

***
### **Hypothyroidism**:
* A life long autoimmune disease with **no cure$^1$**.
* Its cause is very poorly understood, and it is **related to several other nonthyroidal autoimmune diseases** (NSAIDs)$^1$.
* **79%** of predisposition is due to **genetic factors**, allotting **21%** for **environmental** and **sex hormone influences**$^1$. 
*  The major **environmental triggers** that have been identified are$^3$:
    * iodine
    * selenium
    * medications
    * smoking and stress
    * infection
    * sex steroids
    * pregnancy
    * fetal microchimerism
    * radiation exposure
* Often misdiagnosed with other diseases as it can present **non-specific symptoms**$^2$.
* A condition in which **serum thyroid hormones are decreased**$^2$.
* Diagnosed by **anti-thyroid tests** such as *anti-thyroid peroxidase antibody (TPO) and anti-thyroglobulin antibody (TgAb)*$^2$.
* The key (for diagnosis) is to follow up on the levels of thyroid hormone. 
* Some patients may develop lymphoma, and thus a regular examination of the neck area is highly recommended$^1$.

* The most **common laboratory findings** demonstrate a combination of$^1$:  
    1. elevated thyroid-stimulating hormone (TSH)
    2. low thyroxine (T4) levels
    3. increased antithyroid peroxidase (anti-TPO) antibodies
    

* There may also be$^1$:
    1. **Anemia** is present in 30% to 40%.
    2. Decreased glomerular filtration rate (GFR), renal plasma flow, and renal free water clearance with resultant **hyponatremia**.
    3. Elevated **Creatine kinase**.
    4. Increased **Prolactin**.
    5. Elevated **total cholesterol**, **LDL**, and **triglyceride levels**.

* **Primary** (most common) vs **secondary form**$^3$.
* **Grading system** vs. **clinical and subclinical** methods for hypothyroidism classification$^3$.

* Different "most common" causes$^1$:

## Symptoms

**General$^1$**:
1. The **rate of hair growth slows**, and hair can be dry, coarse, dull, and brittle. Diffuse or partial alopecia is not uncommon$^1$.
2. **Myxedema** is an edema-like skin condition (uncommon and only occurs in severe cases)$^1$. 
3. **Scaly and dry skin**, especially on the extensor surfaces, palms, and soles.
4. **Bradycardia** &#8594 decreased thyroid function can increase peripheral vascular resistance by as much as 50% to 60% and **reduce cardiac output** by as much as 30% to 50%.
5. **Fatigue**, **exertional dyspnea**, and **exercise intolerance** are likely associated with a combination of limited pulmonary and cardiac reserve in addition to **decreased muscle strength** or **increased muscle fatigue**.

**Early symptoms$^1$**:
1. Constipation
2. Fatigue
3. Dry skin
4. Weight gain.

**Advanced symptoms$^1$**: 
1. Cold intolerance
2. Decreased sweating 
3. Nerve deafness
4. Peripheral neuropathy
5. Decreased energy
6. Depression
7. Dementia
8. Memory loss
9. Muscle cramps
10. Joint pain
11. Hair loss
12. Apnea
13. Menorrhagia
14. Pressure symptoms in the neck

**Physical symptoms$^1$** :

* Cold and dry skin
* Facial edema particularly periorbital, as well as nonpitting edema involving the hands and feet
* Brittle nails
* Bradycardia
* The delayed relaxation phase of tendon reflexes
* Elevated blood pressure
* Slow speech
* Ataxia
* Macroglossia
* Accumulation of fluid in the pleural and pericardial cavities rarely
* Myxoedema coma (the most severe clinical presentation)

* The various **nonspeciﬁc symptoms of hypothyroidism may not manifest simultaneously**, resulting in a subclinical rate higher than that of hyperthyroidism. 

* In addition, patients with hypothyroidism, such as those with **Hashimoto’s thyroiditis**, are **dependent upon long-term levothyroxine treatment**, which may **affect the manifestations of routine laboratory ﬁndings**.

## Application of Data Science Approaches to Investigate Autoimmune Thyroid Disease in Precision Medicine$^3$


* Data Science can be important for **treating variability** in autoimmune disorders.
* Through ML it is very possible to obtain patterns within patient data, the exploitation of these patterns helps to predict and treat patients in order to **improve clinical disease management**.

> ref 52 for further info.


**Selection algorithms** such as **Kruskal-Wallis’ 
analysis**, **Fisher’s discriminant ratio**, and **Relief-F** have been used to analyze databases containing clinical features (e.g. U.S. Surveillance Epidemiology and End Results (SEER) database) from identified thyroid disease patients.

> ref 51 for further info.

**Data mining** algorithms such as **kNNs, SVMs, ID3ara, Naïve bayes** have been used on health record data sets to analyze factors contributing to autoimmune diseases such as those associated with thyroid disease and help to **find better treatments** for thyroid patients. 

> ref 53, 54 for further info.

## Prediction of LT4 treatment trend for patients suﬀering from hypothyroidism$^4$.

1. The $1^{st}$ paper to focus on the thyroid disease **treatment prediction**.
2. **New set of features** identified on the basis of the endocrinologist's experience in the patient's treatment.
3. **Integrated dataset** build from patient's medical history and current state.

### Dataset Cleaning Methodology
*All 4 datasets (D, D1, D2, D3) described below were validated and tested.*

1. **Original dataset (D)**
    1. Patiens with not much data discarded
    2. Final dataset contained 247 patients (196 female, 51 men), mean age of 46 years 

2. Three macro-groups of pathologies:
    1. Congenital hypothyroidism
    2. Hashimoto’s thyroiditis with hypothyroidism
    3. Hypothyroidism

3. **Dataset 1 (D1)** 
    1. Spline Interpolation for eliminating empty cells
    2. Balancing using Synthetic Minority Oversampling Technique (SMOTE) for mitigating the imbalance problem on D

4. **Dataset 2 (D2)**
    1. Discretization (distributing values into intervals and then replacing them with a certain label) on D


5. **Dataset 3 (D3)** 
    1. Mean Normalization (set values to a common scale) on D

### Model

1. **Data**  
    1. 27 attributes (selected by an expert), LT4 treatment trend (target) which can take 4 values:
        1. Increased
        2. Decreased
        3. Stable
        4. Other (e.g. suspended)

**Classifiers**
1. **Boosting algorithms** 
  
    1. AdaBoost
    2. Gradient Boosting
    3. XGBC
    4. CatBoost

*These approaches use decision trees as weak learners, are not parametric, and do not assume or require data to follow a particular distribution.* 

2. **Decision trees**

    1. Decision Tree
    2. **ExtraTree** (best overall performance)
    3. Random Forest
    
3. **Common approaches**

    1. Naive Bayes
    2. K-Nearest Neighbors
    3. MLP
    
*Exact hyperparameters for each are described in detail*

### Validation

* k-fold cross-validation
* accuracy, precision, recall, F-score
